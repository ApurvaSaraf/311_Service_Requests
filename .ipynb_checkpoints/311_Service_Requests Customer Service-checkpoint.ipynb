## Order the complaint types based on the average ‘Request_Closing_Time’, grouping them for different locations.

```python
nyc_new['Request_Closing_Sec'] = nyc_new['Request_Closing_Time'].astype('timedelta64[s]')
nyc_new['Request_Closing_hr'] = nyc_new['Request_Closing_Time'].astype('timedelta64[h]')+1
nyc_new.head()
```

```python
## Order the complaint types based on the average ‘Request_Closing_Time’, grouping them for different locations.
dd=nyc_new.groupby(['City','Complaint Type']).Request_Closing_hr.mean().unstack()
dd
```

## Perform a statistical test for the following:



```python
## Whether the average response time across complaint types is similar or not (overall)
d=nyc_new.groupby(['Complaint Type',]).Request_Closing_hr.mean().sort_values(ascending=False).round(1)
d
```

```python
from scipy import stats
one_sample_data =d.values
data =d.values.mean()
data
ttest,p_val = stats.ttest_1samp(one_sample_data, 20)

#print("The t-statistic is %.3f and the p-value is %.3f." % one_sample)
```

```python
if p_val<0.05:
    print('Null hypothesis is rejected since p value ({}) is less than 0.05')
else:
    print('Null hypothesis is accepted since p value ({}) is greater than 0.05')
```

```python
## Are the type of complaint or service requested and location related?
loc=nyc_new.groupby(['Complaint Type','City']).Request_Closing_hr.mean().unstack().round(1)
loc
```

```python
loc['ARVERNE']
```

```python
loc['ASTORIA']
```

```python
one_sample_data1 =loc['ARVERNE']
one_sample_data2 =loc['ASTORIA']

_,p_val = stats.ttest_rel(one_sample_data1, one_sample_data2)

#print("The t-statistic is %.3f and the p-value is %.3f." % one_sample)
if p_val<0.05:
    print('Null hypothesis is rejected since p value ({}) is less than 0.05')
else:
    print('Null hypothesis is accepted since p value ({}) is greater than 0.05')

```

```python
f_val,p_val = stats.f_oneway(one_sample_data1,one_sample_data2)
print('F-statistic is =',f_val)
print('p value is =',np.around(p_val,decimals=2))
```
