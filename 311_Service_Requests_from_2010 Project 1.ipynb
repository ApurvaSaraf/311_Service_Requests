```python
# II Perform basic data exploratory analysis:
 #Utilize missing value treatment
nyc=nyc.drop(columns=['Unique Key','Agency','Agency Name','Incident Zip','Incident Address','Street Name','Cross Street 1','Cross Street 2','Intersection Street 1','Intersection Street 2','Address Type','Landmark','Facility Type','Due Date','Resolution Description','Resolution Action Updated Date','Community Board','X Coordinate (State Plane)','Y Coordinate (State Plane)','Park Facility Name','Park Borough','School Name','School Number','School Region','School Code','School Phone Number','School Address','School City','School State','School Zip','School Not Found','School or Citywide Complaint','Vehicle Type','Taxi Company Borough','Taxi Pick Up Location','Bridge Highway Name','Bridge Highway Direction','Road Ramp','Bridge Highway Segment','Garage Lot Name','Ferry Direction','Ferry Terminal Name','Location'],axis=1)
nyc.info()
```

```python
nyc.isnull().sum()
```

```python
nyc['Descriptor']=nyc['Descriptor'].fillna(nyc['Descriptor'].mode()[0])
```

```python
nyc['Location Type']=nyc['Location Type'].fillna(nyc['Location Type'].mode()[0])
nyc['City']=nyc['City'].fillna(nyc['City'].mode()[0])
```

```python
nyc['Latitude'] = nyc['Latitude'].nunique(dropna=True)
nyc['Longitude'] = nyc['Longitude'].nunique(dropna=True)
```

```python
nyc.isnull().sum()
```

```python
#Analyze the date column and remove the entries if it has an incorrect timeline
#Read or convert the columns ‘Created Date’ and Closed Date’ to datetime datatype and create a new column ‘Request_Closing_Time’ as the time elapsed between request creation and request closing. (Hint: Explore the package/module datetime)
nyc['Created Date'] = pd.to_datetime(nyc['Created Date'])
nyc['Closed Date']= pd.to_datetime(nyc['Closed Date'])
```

```python
nyc_new=nyc[~nyc['Closed Date'].isnull()]
nyc_new
```

```python
nyc_new['Request_Closing_Time']=nyc_new['Closed Date']-nyc_new['Created Date']
nyc_new
```

```python
# Identify significant variables by performing a statistical analysis using p-values and chi-square values (Optional)
from scipy.stats import f_oneway
```

```python
stat, p = f_oneway(nyc_new['Noise - Street/Sidewalk'],nyc_new['Blocked Driveway'],nyc_new['Illegal Parking'],nyc_new['Derelict Vehicle'], nyc_new['Noise - Commercial'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Same distributions (fail to reject H0)')
else:
    print('Different distributions (reject H0)')
```
